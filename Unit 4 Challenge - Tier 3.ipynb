**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Calculate the ratio of the Average Price between 1998 and 2018
def create_price_ratio(d):
    price_1998 = float(d['Average_price'][d['year']==1998])
    price_2018= float(d['Average_price'][d['year']==2018])

    if price_1998 == 0 or pd.isna(price_1998) or pd.isna(price_2018):
        return None  ## Avoid division by zero or missing data
    ##I flipped the ratio so it's 2018 / 1998, because that’s usually more intuitive (“how many times higher are prices now”)
    ratio_1998_2018=price_2018/price_1998      
    return ratio_1998_2018
    
```

```python
create_price_ratio(prop_new_group[prop_new_group['London_Borough']=='Barking & Dagenham'])
```

```python
# let's make an empty dictionary, called ratio_final, where we'll store our ratios for each unique London_Borough.
ratio_final={}
# Now let's declare a for loop that will iterate through each of the unique elements of the 'London_Borough' column of our DataFrame prop_new_group.
# Call the iterator variable 'bh'
for bh in prop_new_group['London_Borough'].unique():
    borough_data = prop_new_group[prop_new_group['London_Borough'] == bh]
    ratio_final[bh] = create_price_ratio(borough_data)
# Convert to DataFrame for nicer display
ratio_df = pd.DataFrame(ratio_final.items(), columns=['London_Borough', 'Price_Ratio_2018'])
ratio_df= ratio_df.sort_values(by='Price_Ratio_2018', ascending=False)
ratio_df_top15=ratio_df.head(15)

print(ratio_df_top15)


```

```python
plt.figure(figsize=(12, 6))
bars = plt.bar(ratio_df_top15['London_Borough'], ratio_df_top15['Price_Ratio_2018'], 
               color='lightblue', edgecolor='black')

# Add value labels on top
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, 
             f"{height:.2f}", ha='center', va='bottom', fontsize=8)

plt.xticks(rotation=75, ha='right')
plt.ylabel('Price Ratio (2018 / 1998)', fontsize=12)
plt.title('Increase in Average House Prices in London Boroughs (2018 vs 1998)', 
          fontsize=14, weight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
